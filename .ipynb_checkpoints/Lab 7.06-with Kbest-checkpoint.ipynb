```python
# I will drop RFA_2 since there is only 1 unique value, and unnamed column
categorical = categorical.drop(columns=['RFA_2R'], axis=1)
```

```python
# there are almost 50 clusters, I created another column 'other' for one-hot-encoding
# clusters below 2500 will be called other, in the end I have 13 values to one-hot encode

df = pd.DataFrame(categorical['CLUSTER'].value_counts()).reset_index()

df.columns = ['cluster', 'count']
other_clusters = list(df[df['count']<2500]['cluster'])

def clean_cluster(x):
    if x in other_clusters:
        return 'other'
    else:
        return x
```

```python
categorical['CLUSTER'] = list(map(clean_cluster, categorical['CLUSTER']))
```

```python
# going to drop DOB because we have age anyways.
# going to drop first and last donated month-year, I dont think it is relevant
# I will keep months maximum-minimum donations, maybe it will help with pattern
categorical = categorical.drop(columns=['ODATEW_YR', 'ODATEW_MM', 'DOB_YR',
       'DOB_MM', 'MINRDATE_YR', 'MAXRDATE_YR',
       'LASTDATE_YR', 'MAXRDATE_MM','MINRDATE_MM', 'FIRSTDATE_YR', 'FIRSTDATE_MM','DATASRCE'], axis=1)
```

```python
categorical = categorical.applymap(str)
```

### undersampling - X-y split

I will undersample my data, but first do train-test split and undersample the train set

```python
# X/y split
X = data.drop('TARGET_B',axis = 1)
y = data['TARGET_B']
```

```python
# train-test-split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0) # default fraction is .2
```

downsampling train set

```python
X_train_imb = pd.DataFrame(X_train)
```

```python
trainset = pd.concat((X_train, y_train), axis=1)
```

```python
trainset['TARGET_B'].value_counts()
```

```python
# trainset.head()
```

```python
from sklearn.utils import resample

category_0 = trainset[trainset['TARGET_B'] == 0]
category_1 = trainset[trainset['TARGET_B'] == 1]
```

```python
category_0_undersampled = resample(category_0, 
                                   replace=False, 
                                   n_samples = len(category_1))
```

```python
print(category_0_undersampled.shape)
print(category_1.shape)
```

```python
train_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)
```

```python
# train_downsampled.head()
```

```python
train_downsampled['TARGET_B'].value_counts()
```

now X-y split again to my downsized train set

```python
X_train = train_downsampled.drop('TARGET_B',axis = 1)
y_train = train_downsampled['TARGET_B']
```

```python
#splitting into categorical and numerical
X_train_num = X_train.select_dtypes(include = np.number)
X_train_cat = X_train.select_dtypes(include = object)

X_test_num = X_test.select_dtypes(include = np.number)
X_test_cat = X_test.select_dtypes(include = object)
```

```python
#scale numericals
from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(X_train_num) 
numericals_train_scaled = transformer.transform(X_train_num)
numericals_test_scaled = transformer.transform(X_test_num)
```

```python
#encode categoricals
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='error', drop='first').fit(X_train_cat)
categoricals_train_encoded = encoder.transform(X_train_cat).toarray()
categoricals_test_encoded = encoder.transform(X_test_cat).toarray()
```

```python
X_train_transformed = np.concatenate((numericals_train_scaled,categoricals_train_encoded),axis=1)
X_test_transformed = np.concatenate((numericals_test_scaled,categoricals_test_encoded),axis=1)
```

### building a logistic regression model

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_transformed, y_train)
LR.score(X_test_transformed, y_test)   # LR score is 0.80, it is good number
```

```python
pred = LR.predict(X_test_transformed)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

```python
#while accuracy is not absolutely terrible, a closer look reveals some serious problems
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score


print("precision: ",precision_score(y_test,pred))    # true pos / true pos+false pos
print("recall: ",recall_score(y_test,pred))    #recall is most important, true pos/true pos + false neg
print("f1: ",f1_score(y_test,pred))
```
